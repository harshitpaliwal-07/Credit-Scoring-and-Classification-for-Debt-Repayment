```python
# shape and data types of the data
print(data.shape)
print(data.dtypes)
data.columns = data.columns.str.replace(' ', '') 
data_numeric = data.select_dtypes(include=[np.number])
numeric_cols = data_numeric.columns.values
print(numeric_cols)
data['payment']=data['payment'].replace(.00,0)
data['totpayment']=data['totpayment'].replace(.00,0)
```

```python
print(ext_data.shape)
print(ext_data.dtypes)
ext_data.columns = ext_data.columns.str.replace(' ', '')
ext_data_numeric = ext_data.select_dtypes(include=[np.number])
ext_numeric_cols = ext_data_numeric.columns.values
print(ext_numeric_cols)
```

```python
for col in data.columns:
    pct_missing = np.mean(data[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
for col in ext_data.columns:
    pct_missing = np.mean(ext_data[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
data1=data.merge(ext_data, left_on='ID1',right_on='ID1',how='outer', indicator=True,suffixes=('', '_DROP')).filter(regex='^(?!.*_DROP)').query('_merge == "left_only"').drop(['_merge'],axis=1)
```

```python
new_data=pd.concat([ext_data,data1])
```

```python
cols_to_drop = ['pattype','fintype','lastname','firstname','flg1','flg2','flg3','zipcode','score1','score2','usrid','homephone','workphone','T#PIF','insurance','TPlAct','TPyAct','OLTrn']
final_data = new_data.drop(cols_to_drop, axis=1)
```

```python

print(final_data)

final_data = final_data.astype({"comp":np.int64,"linknumber":np.int64,"casenumber":np.int64,"dateplaced":np.int64,"acPIF":np.int64,"dofb":np.int64,"accountcat":np.int64,"nocontacts":np.int64,"cosigner":np.int64,"lastpayment":np.int64,"servicedate":np.int64,"paydate":np.int64,"clnt":np.int64,"groupcode":np.int64,"score3":np.int64,"priorinspaymt":np.int64,"priorpatpaymt":np.int64,"recovperc":np.int64})
final_data.info()
```

```python
final_data['dateplaced']=pd.to_datetime(final_data['dateplaced'].astype(str), format='%Y%m%d')
final_data['dofb']=final_data['dofb'].replace(0, 19000101)
```

```python
final_data['dofb'] = final_data['dofb'].apply(str)
```

```python
searchfor = ['0000']
final_data["TrueFalsedofb"] = final_data['dofb'].apply(lambda x: 1 if any(i in x for i in searchfor) else 0)
```

```python
final_data["TrueFalsedofb"].sum()   ### DOFB which are wrong
```

```python
final_data['dofb']=pd.to_datetime(final_data['dofb'], format='%Y/%m/%d',errors ="coerce")
```

```python
final_data['paydate']=final_data['paydate'].replace(0, 19000101)
final_data['paydate']=pd.to_datetime(final_data['paydate'].astype(str), format='%Y%m%d')
```

```python
final_data['lastpayment']=final_data['lastpayment'].replace(0, 10100)
final_data['servicedate']=final_data['servicedate'].replace(0, 10100)
```

```python
#add leading zeroes
final_data['lastpayment'] = final_data['lastpayment'].apply(lambda x: '{0:0>6}'.format(x))
final_data['servicedate'] = final_data['servicedate'].apply(lambda x: '{0:0>6}'.format(x))
```

```python
final_data['lastpayment']=pd.to_datetime(final_data['lastpayment'].astype(str), format='%m%d%y')
final_data['servicedate']=pd.to_datetime(final_data['servicedate'].astype(str), format='%m%d%y')
```

```python
final_data['SSN'] = final_data['SSN'].fillna(0)
final_data['zip'] = final_data['zip'].fillna(0)
```

```python
final_data['RLTrn'].replace('\/', '', regex=True, inplace=True)
final_data['RLTrn']=pd.to_datetime(final_data['RLTrn'].astype(str), format='%m%d%y')
```
